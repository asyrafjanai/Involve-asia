f) Clean text and generate new text from Job Responsibilities column: The new text shall not contain any stop words, and the plural words shall be converted into singular words.

```python
from nltk.corpus import stopwords
stop = stopwords.words('english')

job_post['JobDescription'] = job_post['JobDescription'].astype(str)

job_post['JobDescription'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))
```

```python
from textblob import TextBlob
words = "cat dog child goose pants"
blob = TextBlob(words)
plurals = [word.pluralize() for word in blob.words]
print(plurals)
```

```python
for i in range(len(job_post['JobDescription'])):
    words = job_post['JobDescription'][i]
    blob = TextBlob(words)
    single = [word.singularize() for word in blob.words]
    job_post['JobDescription'][i] = single
```

```python
import difflib as diff

#Create a new column 'similarity' and get the result of comparision to it
text_data['similarity'] = text_data.apply(lambda x: diff.SequenceMatcher(None, x[0].strip(), x[1].strip()).ratio(), axis=1) 
```
